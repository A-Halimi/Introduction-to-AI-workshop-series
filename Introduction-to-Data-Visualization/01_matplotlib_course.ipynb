## 3. Basic Plots

### 3.1. Line Chart


**Introduction and Use-Cases:**  
Line charts, sometimes called time series charts, show a series of data points connected by a line. They are one of the most commonly used charts because they are simple but versatile. Line charts are particularly useful for visualizing data that has a sequential relationship, like stock prices over time or a company's growth.




**Basic line chart:**  
To start with, we'll create a basic line chart using synthetic data. We'll plot a sine wave, which is a mathematical curve that describes a smooth periodic oscillation


````python
# ---- Basic Line Chart ----
# Generate synthetic data for the line chart
# We use np.linspace to generate 100 evenly spaced numbers between 0 and 2*pi (which is one full cycle of a sine wave)
x = np.linspace(0, 2 * np.pi, 100)

# Create a new figure with a specified size
plt.figure(figsize=(7, 5))

# Compute the sine of each of these numbers. This will give us the y-coordinates of our data points.
y = np.sin(x)

# Create a basic line chart. 
# The plt.plot function is used to plot the data. 
# The first argument is the x-coordinates and the second argument is the y-coordinates.
plt.plot(x, y)

# Set the title for the chart and labels for x and y axes.
plt.title('Basic Line Chart of a Sine Wave')
plt.xlabel('X Axis (0 to 2π)')
plt.ylabel('Y Axis (Sine values)')

# Display the chart. This will show the line chart of a sine wave.
plt.show()
````


**Customizing the line chart:**  
Customization can make charts more informative and appealing. We'll explore changing colors, adding markers, gridlines, and other enhancements
.

````python
# Import necessary libraries
import numpy as np
import matplotlib.pyplot as plt

# Generate synthetic data for the line chart
x = np.linspace(0, 2 * np.pi, 100)
y = np.sin(x)

# Create a new figure with a specified size
plt.figure(figsize=(15, 7))

# ---- 1. Basic Line Chart ----
plt.subplot(2, 3, 1)
plt.plot(x, y)
plt.title('1. Basic Line Chart')
plt.xlabel('X Axis (0 to 2π)')
plt.ylabel('Y Axis (Sine values)')

# ---- 2. Dashed Line Chart ----
plt.subplot(2, 3, 2)
plt.plot(x, y, 'b--')
plt.title('2. Dashed Line Chart')
plt.xlabel('X Axis (0 to 2π)')
plt.ylabel('Y Axis (Sine values)')

# ---- 3. Line Chart with Markers ----
plt.subplot(2, 3, 3)
plt.plot(x, y, 'bo-')
plt.title('3. Line Chart with Markers')
plt.xlabel('X Axis (0 to 2π)')
plt.ylabel('Y Axis (Sine values)')

# ---- 4. Line Chart with Gridlines ----
plt.subplot(2, 3, 4)
plt.plot(x, y)
plt.grid(True)
plt.title('4. Line Chart with Gridlines')
plt.xlabel('X Axis (0 to 2π)')
plt.ylabel('Y Axis (Sine values)')

# ---- 5. Line Chart with Custom Color and Linewidth ----
plt.subplot(2, 3, 5)
plt.plot(x, y, color='green', linewidth=5)
plt.title('5. Line Chart with Custom Color and Linewidth')
plt.xlabel('X Axis (0 to 2π)')
plt.ylabel('Y Axis (Sine values)')

# ---- 6. Line Chart with an Inset ----
plt.subplot(2, 3, 6)
plt.plot(x, y, 'b-')

# Add an inset (a subplot within a subplot)
# Define the position of the inset: x-position, y-position, width, height
ax_inset = plt.axes([0.95, 0.35, 0.15, 0.15])
ax_inset.plot(x, y, 'b-')
ax_inset.set_xlim(2, 3.4)  # X range around pi
ax_inset.set_ylim(0.5, 1)   # Y range to show the peak
ax_inset.set_title('Inset showing peak around π')

# Main plot title and labels
plt.title('6. Line Chart with an Inset')
plt.xlabel('X Axis (0 to 2π)')
plt.ylabel('Y Axis (Sine values)')

# Adjust the layout for better presentation
plt.tight_layout()

# Adjust the spacing between the subplots
#plt.subplots_adjust(wspace=0.4, hspace=0.3)

# Display the figure with all subplots
plt.show()
````

### 3.2. Bar Chart


**Introduction and Use-Cases:**  
Bar charts are one of the most common types of visualization. They are used to display and compare the number, frequency, or other measure (like mean) for different discrete categories or groups. Bar charts can be horizontal or vertical. In essence, they are used when you have qualitative data and you want to see the size of each category compared to the others.



**Basic bar chart:**  
We'll start by creating a basic vertical bar chart using synthetic data. The data will represent the sales of a product in different months

````python
# ---- Basic Bar Chart ----
# Let's consider a hypothetical scenario where we have sales data of a product for different months.
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May']
sales = [100, 150, 120, 180, 140]

# Using plt.bar() to create a bar chart
plt.bar(months, sales)

# Setting title and labels
plt.title('Product Sales from Jan to May')
plt.xlabel('Month')
plt.ylabel('Sales (in units)')

# Display the basic bar chart
plt.show()
````

**Customizing the bar chart:**  
Aesthetic and informative customizations can make charts more engaging. We'll look into changing colors, adding labels, and adjusting other attributes to make the chart more insightful.


````python
# Data for the bar chart
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May']
sales = [100, 150, 120, 180, 140]
colors = ['blue', 'green', 'red', 'purple', 'cyan']

# Create a new figure with a specified size
plt.figure(figsize=(14, 5))

# ---- Vertical Bar Chart ----
# Add a subplot to the current figure (1 row, 2 columns, first plot)
plt.subplot(1, 2, 1)

# Create a vertical bar chart
bars = plt.bar(months, sales, color=colors, alpha=0.7, edgecolor='black', linewidth=1.2)

# Adding a horizontal grid for better readability
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Setting title and labels
plt.title('Vertical Bar Chart')
plt.xlabel('Month')
plt.ylabel('Sales (in units)')

# Add legend using the bars and months as labels
plt.legend(bars, months)

# ---- Horizontal Bar Chart ----
# Add another subplot to the current figure (1 row, 2 columns, second plot)
plt.subplot(1, 2, 2)

# Create a horizontal bar chart
bars_horizontal = plt.barh(months, sales, color=colors, alpha=0.7, edgecolor='black', linewidth=1.2)

# Adding a vertical grid for better readability
plt.grid(axis='x', linestyle='--', alpha=0.7)

# Setting title and labels
plt.title('Horizontal Bar Chart')
plt.xlabel('Sales (in units)')
plt.ylabel('Month')

# Add legend using the bars_horizontal and months as labels
plt.legend(bars_horizontal, months, loc='best')

# Adjust the layout to ensure there's no overlap and everything fits well
plt.tight_layout()

# Display the figure with both subplots
plt.show()
````

### 3.3. Scatter Plot

**Introduction and Use-Cases:**  
A scatter plot is a type of data visualization that uses dots to represent the values obtained for two different variables - one plotted along the x-axis and the other plotted along the y-axis. Scatter plots are used:

* To observe and show relationships between two numeric variables.
* To identify correlations between variables. For instance, a positive slope (upward trend) suggests a positive correlation.
* To identify outlier points that don't fit the general trend.

**Basic scatter plot:**  

````python
# Sample data for scatter plot
x = np.random.rand(50)
y = x + np.random.normal(0, 0.1, 50)  # y is approximately proportional to x, with some added noise

# Create a basic scatter plot
plt.scatter(x, y)
plt.title('Basic Scatter Plot')
plt.xlabel('X values')
plt.ylabel('Y values')
plt.show()
````

**Customizing scatter plots:**

Scatter plots in Matplotlib are highly customizable. Here, we'll explore various customizations using subplots:

* Changing the color and size of points.
* Adding a regression line to the plot.
* Adjusting the opacity of points.

````python
# Create a new figure with a specified size
plt.figure(figsize=(15, 5))

# 1. Scatter plot with custom color and size
plt.subplot(1, 3, 1)
sizes = np.random.rand(50) * 100  # Random sizes for the points
plt.scatter(x, y, color='red', s=sizes)
plt.title('Scatter Plot with Custom Color and Size')
plt.xlabel('X values')
plt.ylabel('Y values')

# 2. Scatter plot with a regression line
plt.subplot(1, 3, 2)
plt.scatter(x, y)
m, b = np.polyfit(x, y, 1)  # Compute the slope and y-intercept of the regression line
plt.plot(x, m*x + b, color='red')  # Plot the regression line
plt.title('Scatter Plot with Regression Line')
plt.xlabel('X values')
plt.ylabel('Y values')

# 3. Scatter plot with adjusted opacity
plt.subplot(1, 3, 3)
plt.scatter(x, y, alpha=0.5)  # The alpha parameter controls the opacity
plt.title('Scatter Plot with Adjusted Opacity')
plt.xlabel('X values')
plt.ylabel('Y values')

# Adjust the layout for better presentation
plt.tight_layout()

# Display the figure with all subplots
plt.show()
````

### 3.4. Pie Chart

**Introduction and Use-Cases:**

A pie chart is a circular statistical graphic, which is divided into slices to illustrate numerical proportions. Each slice of the pie represents a category and is sized by its corresponding quantity.

Use-cases for pie charts include:

* Displaying market share of companies.
* Representing election results.
* Visualizing the distribution of sales categories.
* Understanding the composition of a dataset in terms of categories or labels.

**Basic pie chart**

````python
# Sample data for pie chart
labels = ['A', 'B', 'C', 'D']
sizes = [15, 30, 45, 10]

# Create a basic pie chart
plt.figure(figsize=(6,6))
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90) ##  autopct='%1.1f%%' is used to label the wedges with their numeric value, startangle=90 rotates the start of the pie chart to vertical.
plt.title('Basic Pie Chart')
plt.show()
````

**Customizing pie charts:**

Pie charts in Matplotlib are highly customizable. Here, we'll explore various customizations using subplots:

* Pie chart with custom colors.
* Exploding slices (pulling out slices).
* Adding a shadow for a 3D effect.

````python
# Create a new figure with a specified size
plt.figure(figsize=(15, 5))

# 1. Pie chart with custom colors
colors = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue']
plt.subplot(1, 3, 1)
plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=90)
plt.title('Pie Chart with Custom Colors')

# 2. Pie chart with exploding slices
explode = (0.1, 0, 0, 0)  # Only 'A' will explode out
plt.subplot(1, 3, 2)
plt.pie(sizes, labels=labels, explode=explode, autopct='%1.1f%%', startangle=90)
plt.title('Pie Chart with Exploding Slices')

# 3. Pie chart with shadow
plt.subplot(1, 3, 3)
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90, shadow=True)
plt.title('Pie Chart with Shadow')

# Adjust the layout for better presentation
plt.tight_layout()

# Display the figure with all subplots
plt.show()
````

### 3.5. Histogram 


**Introduction and use-cases**

A histogram is an approximate representation of the distribution of numerical or categorical data. It is an estimate of the probability distribution of a continuous variable. To construct a histogram, the first step is to "bin" the range of values—that is, divide the entire range of values into a series of intervals—and then count how many values fall into each interval.

Use-cases for histograms include:

* Understanding the distribution of scores in an exam.
* Visualizing the age distribution of a population.
* Examining the distribution of prices in a product category.

````python
# Sample data for histogram: 1000 random numbers drawn from a normal distribution
data = np.random.randn(1000)

# Create a basic histogram
plt.figure(figsize=(7,5))
plt.hist(data, bins=30)  ## The bins parameter determines how many bins (intervals) the data will be divided into. Each bin will represent a frequency of values within a range.
plt.title('Basic Histogram')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()
````

**Customizing histograms:**

Histograms in Matplotlib offer various customization options. Here, we'll explore:

* Histogram with Custom Bin Size and Color: This plot showcases a histogram with 50 bins and a sky-blue color.
* Cumulative Histogram: Here, the histogram is cumulative, meaning each bin contains the sum of all the previous bins. This kind of histogram provides a view of the data distribution.
* Histogram with a Density Curve: This plot showcases a histogram overlaid with a density curve. The curve provides a smoothed version of the histogram and is particularly helpful in understanding the distribution's underlying shape.



````python
# Re-generate the sample data for histogram: 1000 random numbers drawn from a normal distribution
data = np.random.randn(1000)

# Create a new figure with a specified size
plt.figure(figsize=(15, 5))

# 1. Histogram with custom bin size and color
plt.subplot(1, 3, 1)
plt.hist(data, bins=50, color='skyblue', edgecolor='black')
plt.title('Histogram with Custom Bin Size and Color')
plt.xlabel('Value')
plt.ylabel('Frequency')

# 2. Cumulative histogram
plt.subplot(1, 3, 2)
plt.hist(data, bins=30, cumulative=True)
plt.title('Cumulative Histogram')
plt.xlabel('Value')
plt.ylabel('Cumulative Frequency')

# 3. Histogram with a density curve using only Matplotlib
plt.subplot(1, 3, 3)
plt.hist(data, bins=30, density=True, color='lightgrey', alpha=0.7)

# To add a density curve using Matplotlib, we'll compute the kernel density estimate ourselves
density = gaussian_kde(data)
x_vals = np.linspace(min(data), max(data), 1000)
y_vals = density(x_vals)
plt.plot(x_vals, y_vals, color='black')

plt.title('Histogram with Density Curve')
plt.xlabel('Value')
plt.ylabel('Density')

# Adjust the layout for better presentation
plt.tight_layout()

# Display the figure with all subplots
plt.show()
````

### 3.6. Area Chart

**Introduction and use-cases**

An area chart represents the quantity of data over time. It's similar to a line chart, but the area under the line is filled in, which gives it the appearance of an 'area'. This filled area can represent volume or another quantity across a range.

Use-cases for area charts include:

* Visualizing stock market performance over time.
* Representing website traffic or sales over a period.
* Tracking resource usage or capacity.

**Basic area chart**

````python
# Sample data for area chart: a simple linear increase
x = np.linspace(0, 10, 100)
y = x

# Create a basic area chart
plt.figure(figsize=(7,5))
plt.fill_between(x, y)
plt.title('Basic Area Chart')
plt.xlabel('X values')
plt.ylabel('Y values')
plt.show()
````

**Customizing area charts**:

Area charts in Matplotlib are flexible and allow for various customizations. Here, we'll explore:

* Area chart with custom fill color.
* Area chart with multiple fills.
* Area chart with a baseline.

````python
# Create a new figure with a specified size
plt.figure(figsize=(15, 5))

# 1. Area chart with custom fill color
plt.subplot(1, 3, 1)
plt.fill_between(x, y, color='skyblue')
plt.title('Area Chart with Custom Fill Color')
plt.xlabel('X values')
plt.ylabel('Y values')

# 2. Area chart with multiple fills
y2 = x**2
plt.subplot(1, 3, 2)
plt.fill_between(x, y, color='skyblue', label='Linear')
plt.fill_between(x, y2, color='salmon', label='Quadratic')
plt.title('Area Chart with Multiple Fills')
plt.xlabel('X values')
plt.ylabel('Y values')
plt.legend()

# 3. Area chart with a baseline
plt.subplot(1, 3, 3)
plt.fill_between(x, y2, y, where=(y2 > y), color='skyblue', interpolate=True, label='Above Linear')
plt.fill_between(x, y2, y, where=(y2 <= y), color='salmon', interpolate=True, label='Below Linear')
plt.title('Area Chart with a Baseline')
plt.xlabel('X values')
plt.ylabel('Y values')
plt.legend()

# Adjust the layout for better presentation
plt.tight_layout()

# Display the figure with all subplots
plt.show()
````

* Area Chart with Gradient Fill: This showcases an area chart with a gradient fill effect. The gradient is achieved by overlaying two fills with varying alpha (transparency) values.
    
* Stacked Area Chart: This chart demonstrates the stacking of two datasets. The salmon-colored area represents the second dataset stacked on top of the sky-blue one.

* Area Chart with a Pattern Fill: Here, a hatch pattern is used to fill the area, which can be especially useful for print scenarios where color differentiation might be challenging.
    
These customizations demonstrate the versatility of area charts in Matplotlib, allowing for a wide range of visual effects to suit various presentation needs.

````python
# Create a new figure with a specified size
plt.figure(figsize=(15, 5))

# Sample data
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.sin(x) * 0.5

# 1. Area Chart with Gradient Fill
plt.subplot(1, 3, 1)
plt.fill_between(x, y1, color='skyblue', alpha=0.5, label='Gradient Top')
plt.fill_between(x, 0, y1, color='blue', alpha=0.5, label='Gradient Bottom')
plt.title('Area Chart with Gradient Fill')
plt.xlabel('X values')
plt.ylabel('Y values')
plt.legend()

# 2. Stacked Area Chart
plt.subplot(1, 3, 2)
plt.fill_between(x, y1, color='skyblue', label='Dataset 1')
plt.fill_between(x, y1, y1+y2, color='salmon', label='Dataset 2')
plt.title('Stacked Area Chart')
plt.xlabel('X values')
plt.ylabel('Y values')
plt.legend()

# 3. Area Chart with a Pattern Fill
plt.subplot(1, 3, 3)
plt.plot(x, y1, color='black')  # adding a line for clarity
plt.fill_between(x, y1, hatch='/', color='lightgrey', edgecolor='gray')
plt.title('Area Chart with Pattern Fill')
plt.xlabel('X values')
plt.ylabel('Y values')

# Adjust the layout for better presentation
plt.tight_layout()

# Display the figure with all subplots
plt.show()
````

### 3.7. 3D Plot

**Introduction and use-cases**

3D plots are an extension of our familiar 2D plots, allowing us to visualize three dimensions of data. These are particularly useful when we need to understand the relationship between not two, but three variables.

Use-cases for 3D plots include:

* Representing geographical data, where x and y might be longitude and latitude, and z represents elevation.
* Visualizing multi-dimensional datasets in machine learning and data analytics.
* Graphing mathematical functions in three dimensions.

**Basic 3D scatter plot**

````python
from mpl_toolkits.mplot3d import Axes3D

# Sample data
x = np.random.rand(50)
y = np.random.rand(50)
z = np.random.rand(50)

fig = plt.figure(figsize=(7, 5))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(x, y, z)

ax.set_title('Basic 3D Scatter Plot')
ax.set_xlabel('X values')
ax.set_ylabel('Y values')
ax.set_zlabel('Z values')

plt.show()
````

**Customizing 3D plots**

3D plots in Matplotlib offer a wide range of customization options. Here, we'll explore:


* 3D Line Plot: A simple three-dimensional line plot.
* 3D Bar Chart: A basic example of a 3D bar chart.
* 3D Surface Plot: Visualization of a mathematical function using a surface plot.
* 3D Scatter Plot with Color Mapping: Points are color-mapped based on their z-values.
* 3D Bar Chart (Sales Volume): Represents sales volume in three different regions across 12 months.
* 3D Wireframe Plot: A skeletal representation of the surface plot.

````python
# Correcting the bar colors for the 3D Bar Chart
bar_colors_region = ['blue', 'green', 'red']  # colors for each region

# Create a new figure with a specified size
fig = plt.figure(figsize=(18, 10))

# Parameters for the helix
a = 1  # radius of the helix
b = 0.1  # pitch of the helix

# Generate t values
t = np.linspace(0, 4 * np.pi, 400)  # 400 points over 4 periods of the helix

# Compute x, y, z coordinates
x_line = a * np.cos(t)
y_line = a * np.sin(t)
z_line = b * t

# First Figure

# 1. 3D Line Plot
ax1 = fig.add_subplot(2, 3, 1, projection='3d')
ax1.plot(x_line, y_line, z_line, color='blue')
ax1.set_title('3D Line Plot')
ax1.set_xlabel('X values')
ax1.set_ylabel('Y values')
ax1.set_zlabel('Z values')


# Create a grid of positions for the bars
x1 = np.arange(0, 3)
y1 = np.arange(0, 3)
x_bar, y_bar = np.meshgrid(x1, y1)

# Flatten the grid arrays to get 1D arrays of x, y positions
x_bar = x_bar.ravel()
y_bar = y_bar.ravel()

# Set the z positions of the bars to zero, and their heights (dz) to the values of a function
z_bar = np.zeros_like(x_bar)
dz = np.sin(np.sqrt(x_bar**2 + y_bar**2))

# Set the dimensions of the bars
dx = dy = 0.8  # Width and depth of the bars

# 2. 3D Bar Chart
ax2 = fig.add_subplot(2, 3, 2, projection='3d')
ax2.bar3d(x_bar, y_bar, z_bar, dx, dy, dz, color='salmon')
ax2.set_title('3D Bar Chart')
ax2.set_xlabel('X values')
ax2.set_ylabel('Y values')
ax2.set_zlabel('Z values')

# Generate data for the surface plot
x = np.linspace(-5, 5, 100)  # 100 points between -5 and 5
y = np.linspace(-5, 5, 100)  # 100 points between -5 and 5
x, y = np.meshgrid(x, y)  # Create a grid of (x, y) coordinates
z = np.sin(np.sqrt(x**2 + y**2))  # Compute z values using the function z = sin(sqrt(x^2 + y^2))

# 3. 3D Surface Plot
ax3 = fig.add_subplot(2, 3, 3, projection='3d')
surf = ax3.plot_surface(x, y, z, cmap='viridis')
fig.colorbar(surf, ax=ax3, shrink=0.5, aspect=5)
ax3.set_title('3D Surface Plot')
ax3.set_xlabel('X values')
ax3.set_ylabel('Y values')
ax3.set_zlabel('Z values')

# Second Figure
# Generate synthetic data for the scatter plot
np.random.seed(0)  # Seed the random number generator for reproducibility
n_points = 100
x_scatter = np.random.rand(n_points)
y_scatter = np.random.rand(n_points)
z_scatter = np.random.rand(n_points)

# Generate color values based on the distance of each point from the origin
colors = np.sqrt(x_scatter**2 + y_scatter**2 + z_scatter**2)


# 1. 3D Scatter Plot with Color Mapping
ax4 = fig.add_subplot(2, 3, 4, projection='3d')
scatter = ax4.scatter(x_scatter, y_scatter, z_scatter, c=colors, cmap='viridis')
fig.colorbar(scatter, ax=ax4, shrink=0.6, aspect=5)
ax4.set_title('3D Scatter Plot with Color Mapping')
ax4.set_xlabel('X values')
ax4.set_ylabel('Y values')
ax4.set_zlabel('Z values')

# Synthetic data for demonstration
regions = [0, 1, 2]
region_names = ['Region A', 'Region B', 'Region C']
bar_colors_region = ['blue', 'green', 'red']
months = np.arange(12)  # Months represented as numbers 0-11
sales_volume = {
    0: np.random.randint(50, 100, 12),  # Random sales volume for Region A
    1: np.random.randint(30, 80, 12),   # Random sales volume for Region B
    2: np.random.randint(20, 70, 12)    # Random sales volume for Region C
}


# 2. 3D Bar Chart (Monthly Sales Volume in Three Regions)
ax5 = fig.add_subplot(2, 3, 5, projection='3d')
for region, color, region_name in zip(regions, bar_colors_region, region_names):
    ax5.bar(months, sales_volume[region], zdir='y', zs=region, width=0.6, color=color, alpha=0.7, label=region_name)
    
ax5.set_title('Monthly Sales Volume in Three Regions')
ax5.set_xlabel('Months')
ax5.set_ylabel('Regions')
ax5.set_zlabel('Sales Volume')
ax5.set_yticks(regions)
ax5.set_yticklabels(region_names)
ax5.set_xticks(months)
ax5.set_xticklabels(['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'], rotation=45)

# 3. 3D Wireframe Plot
ax6 = fig.add_subplot(2, 3, 6, projection='3d')
ax6.plot_wireframe(x, y, z, color='black')
ax6.set_title('3D Wireframe Plot')
ax6.set_xlabel('X values')
ax6.set_ylabel('Y values')
ax6.set_zlabel('Z values')

# Adjust the layout for better presentation
plt.tight_layout()

# Display the combined figure
plt.show()
````

**Image with Colorbar and Bicubic Interpolation:**

* This visualization uses the 'hot' colormap.
* A colorbar is added to the side, providing a reference for the data values.
* The 'bicubic' interpolation method is applied. Interpolation methods determine how the data is rendered when displayed at a resolution different from the original. Bicubic interpolation often results in smoother images.

These customizations demonstrate the versatility of image plotting in Matplotlib. Adjusting colormaps and interpolations can greatly influence the perception and understanding of the displayed data.

````python
# Generate a random matrix to simulate grayscale image data
image_data = np.random.random((100, 100))

# Create a figure and axes
fig, axs = plt.subplots(1, 3, figsize=(15, 5))

# ---- Basic Image Plot ----
axs[0].imshow(image_data, cmap='gray')
axs[0].set_title('Basic Grayscale Image')
axs[0].axis('off')  # Hide axes for a cleaner look

# ---- Customization 1: Different Colormap ----
# Here, we're using the 'viridis' colormap instead of grayscale.
axs[1].imshow(image_data, cmap='viridis')
axs[1].set_title('Image with Viridis Colormap')
axs[1].axis('off')

# ---- Customization 2: Adding a Colorbar and Interpolation ----
# We're using the 'hot' colormap and adding a colorbar to the side.
# Additionally, we're applying a 'bicubic' interpolation method.
img = axs[2].imshow(image_data, cmap='hot', interpolation='bicubic')
fig.colorbar(img, ax=axs[2], fraction=0.046, pad=0.04)
axs[2].set_title('Image with Colorbar and Bicubic Interpolation')
axs[2].axis('off')

# Display the figure with all subplots
plt.tight_layout()
plt.show()
````

````python
import matplotlib as mpl

data1, data2, data3, data4 = np.random.randn(4, 100)  # make 4 random data sets

X, Y = np.meshgrid(np.linspace(-3, 3, 128), np.linspace(-3, 3, 128))
Z = (1 - X / 2 + X ** 5 + Y ** 3) * np.exp(-(X ** 2) - Y ** 2)

fig, axs = plt.subplots(2, 2, figsize=(10, 7), layout="constrained")
pc = axs[0, 0].pcolormesh(X, Y, Z, vmin=-1, vmax=1, cmap="RdBu_r")
fig.colorbar(pc, ax=axs[0, 0])
axs[0, 0].set_title("pcolormesh()")

co = axs[0, 1].contourf(X, Y, Z, levels=np.linspace(-1.25, 1.25, 11))
fig.colorbar(co, ax=axs[0, 1])
axs[0, 1].set_title("contourf()")

pc = axs[1, 0].imshow(
    Z ** 2 * 100, cmap="plasma", norm=mpl.colors.LogNorm(vmin=0.01, vmax=100)
)
fig.colorbar(pc, ax=axs[1, 0], extend="both")
axs[1, 0].set_title("imshow() with LogNorm()")
````
